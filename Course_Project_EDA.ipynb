```python
# Convert 'Impact', 'Range', 'RPM', and 'Magazine' columns to numerical data
for column in ['Impact', 'Range', 'RPM', 'Magazine']:
  df[column] = pd.to_numeric(df[column], errors='coerce')
```

```python
import matplotlib.pyplot as plt

# Create a scatter plot with Impact vs. Range, colored by Type
plt.figure(figsize=(10, 6))
for weapon_type in df['Type'].unique():
  subset = df[df['Type'] == weapon_type]
  plt.scatter(subset['Range'], subset['Impact'], label=weapon_type, alpha=0.7)

plt.xlabel('Range')
plt.ylabel('Impact')
plt.title('Impact vs. Range by Weapon Type')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Create a scatter plot with Impact vs. Range, colored by Rarity
plt.figure(figsize=(10, 6))
for rarity in df['Rarity'].unique():
  subset = df[df['Rarity'] == rarity]
  plt.scatter(subset['Range'], subset['Impact'], label=rarity, alpha=0.7)

plt.xlabel('Range')
plt.ylabel('Impact')
plt.title('Impact vs. Range by Weapon Rarity')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Assuming the CSV file is in the same directory as your notebook
df = pd.read_csv('Destiny_Guns.csv')

df = df.iloc[:, 1:]
df

# Convert 'Impact', 'Range', 'RPM', and 'Magazine' columns to numerical data
for column in ['Impact', 'Range', 'RPM', 'Magazine']:
  df[column] = pd.to_numeric(df[column], errors='coerce')


# Create a scatter plot with RPM vs. Magazine, colored by Element
plt.figure(figsize=(10, 6))
for element in df['Element'].unique():
  subset = df[df['Element'] == element]
  plt.scatter(subset['RPM'], subset['Magazine'], label=element, alpha=0.7)

plt.xlabel('RPM')
plt.ylabel('Magazine')
plt.title('RPM vs. Magazine by Element')
plt.legend()
plt.grid(True)
plt.show()
```
